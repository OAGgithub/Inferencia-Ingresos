# Ministerio de la mujer

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ministerio de la mujer
df_mujer = pd.read_excel("NOMINA_Minis_Mujer.xlsx")
```

```python
df_mujer.head()
```

```python
df_mujer.info()
```

```python
df_mujer.columns
```

```python
unwanted_columns = ["NO. ", "OTROS ING. ", "TOTAL ING. ", "OTROS DESC.", "ISR", "TOTAL DESC. ", "ESTATUS "]
desired_columns = [column for column in list(df_mujer.columns) if column not in unwanted_columns]
df_mujer = df_mujer[desired_columns]
```

```python
df_mujer
```

```python
rename_columns = {
    "NOMBRE": "Nombre",
    "FUNCIÓN": "Cargo",
    "DIRECCIÓN": "Departamento",
    "GÉNERO": "Sexo",
    "SUELDO BRUTO (RD$)": "Sueldo bruto",
    "NETO ": "Sueldo neto",
}

df_mujer.rename(columns=rename_columns, inplace=True)
```

```python
df_mujer.isna().sum().sort_values(ascending=False)/len(df_mujer)
```

```python
df_mujer = df_mujer.dropna()
```

```python
df_mujer.columns
```

```python
df_mujer = df_mujer[['Nombre', 'Cargo', 'Departamento', 'Sexo', 'Sueldo bruto', 'AFP', 'SFS', 'Sueldo neto']]
```

```python
df_mujer.info()
```

```python
# determining the name of the file
file_name = 'Clean_NOMINA_Minis_Mujer.csv'

# saving the excel
df_mujer.to_csv(file_name)
print('df_mujer is written to Excel File successfully.')
```

# Ministerio de Administración Pública

```python
# Ministerio de administración pública
df_hacienda = pd.read_excel("NOMINA_Minis_Hacienda.xlsx")
```

```python
df_hacienda.head()
```

```python
df_hacienda.info()
```

```python
df_hacienda.columns
```

```python
unwanted_columns = ["Ubicación", "Tipo Empleado", "Impuesto sobre la renta", "SFS Adicional", "Otros Descuentos", "Seguro Vida INAVI", "Total Descuentos",]
desired_columns = [column for column in list(df_hacienda.columns) if column not in unwanted_columns]
df_hacienda = df_hacienda[desired_columns]
```

```python
df_hacienda.isna().sum().sort_values(ascending=False)/len(df_hacienda)
```

```python
df_hacienda = df_hacienda.dropna()
```

```python
df_hacienda.head()
```

```python
df_hacienda = df_hacienda[['Nombre', 'Cargo', 'Departamento', 'Sexo', 'Sueldo bruto', 'AFP', 'SFS', 'Sueldo neto']]
```

```python
# determining the name of the file
file_name = 'Clean_NOMINA_Minis_Hacienda.csv'

# saving the excel
df_hacienda.to_csv(file_name)
print('df_hacienda is written to CSV File successfully.')
```
